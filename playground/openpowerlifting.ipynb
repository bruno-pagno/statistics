```python
# Chapter 1: Exploratory Data Analysis
import pandas as pd

data = pd.read_csv('../datasets/openpowerlifting.csv')
data
```

```python
bodyweight = data['BodyweightKg']
print('bodyweight mean', bodyweight.mean())
print('bodyweight min', bodyweight.min())
print('bodyweight max', bodyweight.max())
```

```python
categories = data['Division'].unique()
print('number of categories:', len(categories))
print(categories)
```

```python
bench_mean = data['BodyweightKg'].dropna().mean()
squat_mean = data['BestSquatKg'].dropna().mean()
deadlift_mean = data['BestDeadliftKg'].dropna().mean()
print('bench mean:', bench_mean)
print('squat mean:', squat_mean)
print('deadlift mean:', deadlift_mean)
```

```python
bench_max = data['BodyweightKg'].dropna().max()
squat_max = data['BestSquatKg'].dropna().max()
deadlift_max = data['BestDeadliftKg'].dropna().max()
print('bench max:', bench_max)
print('squat max:', squat_max)
print('deadlift max:', deadlift_max)
bench_min = data['BodyweightKg'].dropna().min()
squat_min = data['BestSquatKg'].dropna().min()
deadlift_min = data['BestDeadliftKg'].dropna().min()
print('bench max:', bench_min)
print('squat max:', squat_min, '<- attemp fail or broken data?')
print('deadlift max:', deadlift_min, '<- attemp fail or broken data?')
```
